```python
"""
adding Gaussing distribution to anti proton data array
"""

import numpy as np
import matplotlib.pyplot as plt
import statistics
from numpy import array
import pandas as pd
import random
pd.set_option('display.float_format', lambda x: '%0.4f' % x)

data = []

"""
Read data from total data file
--------------------
"""
print("Reading data ...")
data = array(pd.read_table("/Users/jiancheng/GRAMS/Toy Model/data/cut/TOF_anti_proton_1e7_0_500_0_4_result.txt", delimiter=" "))
print("finish reading data!")

print(type(data[1, 0]))
print(type(data))
print(data)

sigma = 0.3    #energy resolution percentage
deltaT = 0    #timing resolution/ns

time_pbar_err = np.array([None] * len(data[:,0]))
dEOut_pbar_err = np.array([None] * len(data[:,0]))
dEIn_pbar_err = np.array([None] * len(data[:,0]))
pbar = data
for i in range(len(data[:,0])):
    dEOut_pbar_err[i] = random.gauss(data[i, 0], sigma * data[i, 0])
    dEIn_pbar_err[i] = random.gauss(data[i, 1], sigma * data[i, 1])
    time_pbar_err[i] = data[i, 2] + deltaT * 2 * (random.random() - 0.5)
    print("applying error:"+str(int(100*i/len(data[:,0])))+"%", end='\r')

plt.scatter(time_pbar_err, dEOut_pbar_err, marker = '.', color = 'b', label = 'TOF out layer')
plt.scatter(time_pbar_err, dEIn_pbar_err, marker = '.', color = 'r', label = 'TOF inner layer')
plt.title('anti proton TOF with error')
plt.xlabel('Time [ns]')
plt.ylabel('Energy diposit')
plt.legend(loc='lower right')
#plt.axis([0, 20, 0, 10])
plt.grid(alpha = 0.2)
plt.show()
```
