```python
"""
example plot
"""
import corner
import numpy as np

ndim, nsamples = 2, 10000
np.random.seed(42)
samples = np.random.randn(ndim * nsamples).reshape([nsamples, ndim])
np.random.seed(7)
samples2 = 3 + 0.5*np.random.randn(ndim * nsamples).reshape([nsamples, ndim])
figure1 = corner.corner(samples, color='b')
figure2 = corner.corner(samples2, fig=figure1, color='r')
```
