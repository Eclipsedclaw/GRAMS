```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import statistics
from numpy import array
import pandas as pd
import math
import time
from scipy.stats import norm
```

```python

#model A
anti_proton_mu = 2.65 + 3.65 * np.log(2)
anti_deuteron_mu = 2.65 + 3.65 * np.log(4)
anti_He3_mu = 2.65 + 3.65 * np.log(6)

anti_proton_sigma = np.sqrt(anti_proton_mu * 0.174 * math.pow(2, 0.4))
anti_deuteron_sigma = np.sqrt(anti_deuteron_mu * 0.174 * math.pow(4, 0.4))
anti_He3_sigma = np.sqrt(anti_He3_mu * 0.174 * math.pow(6, 0.4))
"""
#model B
anti_proton_mu = 2.65 + 3.65 * np.log(2)
anti_deuteron_mu = 2 * anti_proton_mu
anti_He3_mu = 3 * anti_proton_mu
anti_proton_sigma = np.sqrt(anti_proton_mu * 0.174 * math.pow(2, 0.4))
anti_deuteron_sigma = np.sqrt(2) * anti_proton_sigma
anti_He3_sigma = np.sqrt(3) * anti_proton_sigma
"""
x_axis = np.arange(0, 24, 0.001)

anti_proton_M = norm.pdf(x_axis,anti_proton_mu,anti_proton_sigma)
anti_proton_xmax = x_axis[np.argmax(anti_proton_M)]
anti_proton_ymax = anti_proton_M.max()

#anti_deuteron_M = norm.pdf(x_axis,anti_deuteron_mu,anti_deuteron_sigma)
#anti_deuteron_xmax = x_axis[np.argmax(anti_deuteron_M)]
#anti_deuteron_ymax = anti_deuteron_M.max()

anti_He3_M = norm.pdf(x_axis,anti_He3_mu,anti_He3_sigma)
anti_He3_xmax = x_axis[np.argmax(anti_He3_M)]
anti_He3_ymax = anti_He3_M.max()

plt.figure(figsize=(10, 8))

plt.text(2/3 * anti_proton_xmax, anti_proton_ymax, '<Charged Pion>=%.2f'%(2/3 * anti_proton_xmax), color='r')
plt.plot(2/3 * x_axis, anti_proton_M, color = 'r', label='anti proton')
plt.plot(2/3 * np.array([anti_proton_xmax, anti_proton_xmax]), [0, anti_proton_ymax], '--r', alpha = 0.3)

#plt.text(2/3 * anti_deuteron_xmax, anti_deuteron_ymax, '<Charged Pion>=%.2f'%(2/3 * anti_deuteron_xmax), color='g')
#plt.plot(2/3 * x_axis, anti_deuteron_M, color = 'g', label='anti deuteron')
#plt.plot(2/3 * np.array([anti_deuteron_xmax, anti_deuteron_xmax]), [0, anti_deuteron_ymax], '--g', alpha = 0.3)

plt.text(2/3 * anti_He3_xmax, anti_He3_ymax, '<Charged Pion>=%.2f'%(2/3 * anti_He3_xmax), color='b')
plt.plot(2/3 * x_axis, anti_He3_M, color = 'b', label='anti helium3')
plt.plot(2/3 * np.array([anti_He3_xmax, anti_He3_xmax]), [0, anti_He3_ymax], '--b', alpha = 0.3)

plt.xlabel("number of charged pions")
plt.ylabel(" ")
#plt.yscale('log')
plt.ylim([1e-6, 1])
plt.legend()
plt.grid(alpha=0.2)
plt.title('LArTPC charged pion multiplicity for Model A')
```
