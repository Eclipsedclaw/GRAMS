```python
from scipy.ndimage import gaussian_filter1d

def smooth_with_gaussian(x, y, sigma=1):
    """
    Apply Gaussian kernel smoothing to the data.
    
    Parameters:
    - x: Array of x-values.
    - y: Array of y-values.
    - sigma: The standard deviation of the Gaussian kernel.
    
    Returns:
    - y_smooth: The smoothed y-values.
    """
    y_smooth = gaussian_filter1d(y, sigma=sigma)
    return x, y_smooth


DM_coogan_x, DM_coogan_y = smooth_with_gaussian(DM_coogan_x, DM_coogan_y)
DM_Korsmeier_upper_x, DM_Korsmeier_upper_y = smooth_with_gaussian(DM_Korsmeier_upper_x, DM_Korsmeier_upper_y)
DM_Korsmeier_lower_x, DM_Korsmeier_lower_y = smooth_with_gaussian(DM_Korsmeier_lower_x, DM_Korsmeier_lower_y)
DM_Nan_Li_upper_x, DM_Nan_Li_upper_y = smooth_with_gaussian(DM_Nan_Li_upper_x, DM_Nan_Li_upper_y)
DM_Nan_Li_lower_x, DM_Nan_Li_lower_y = smooth_with_gaussian(DM_Nan_Li_lower_x, DM_Nan_Li_lower_y)
DM_Nan_Li_x, DM_Nan_Li_y = smooth_with_gaussian(DM_Nan_Li_x, DM_Nan_Li_y)
DM_MKachelrieβ_x, DM_MKachelrieβ_y = smooth_with_gaussian(DM_MKachelrieβ_x, DM_MKachelrieβ_y)
DM_M_Winkler_x, DM_M_Winkler_y = smooth_with_gaussian(DM_M_Winkler_x, DM_M_Winkler_y)
BKG_Blum_x, BKG_Blum_y = smooth_with_gaussian(BKG_Blum_x, BKG_Blum_y)
BKG_Poulin_x, BKG_Poulin_y = smooth_with_gaussian(BKG_Poulin_x, BKG_Poulin_y)
BKG_A_Shukla_x, BKG_A_Shukla_y = smooth_with_gaussian(BKG_A_Shukla_x, BKG_A_Shukla_y)
BKG_M_Kachelrie_background_x, BKG_M_Kachelrie_background_y = smooth_with_gaussian(BKG_M_Kachelrie_background_x, BKG_M_Kachelrie_background_y)
```

```python
import matplotlib.pyplot as plt
import numpy as np

GRAMS_antihelium3_x = np.linspace(0.1, 0.5, 100)
GRAMS_antihelium3_y = np.full_like(GRAMS_antihelium3_x, 4.26e-7)
GRAMS_antihelium3_upper = np.full_like(GRAMS_antihelium3_x, 1)
GRAMS_antihelium3_lower = np.full_like(GRAMS_antihelium3_x, 4.26e-7/3.09)

# Interpolate lower y-values onto the upper x-values for both Korsmeier and Nan Li
DM_Korsmeier_lower_y_interp = np.interp(DM_Korsmeier_upper_x, DM_Korsmeier_lower_x, DM_Korsmeier_lower_y)
DM_Nan_Li_lower_y_interp = np.interp(DM_Nan_Li_upper_x, DM_Nan_Li_lower_x, DM_Nan_Li_lower_y)

# Create the plot
plt.figure(figsize=(12, 8))

# GRAMS sensitivity
GRAMS0 = plt.plot(GRAMS_antihelium3_x[0:98], GRAMS_antihelium3_y[0:98], color="#D62728", linewidth = 5, zorder=3)
GRAMS1 = plt.fill_between(GRAMS_antihelium3_x, GRAMS_antihelium3_upper, GRAMS_antihelium3_y, color="#D62728", alpha=0.7, label="GRAMS sensitivity", zorder=2)
GRAMS2 = plt.fill_between(GRAMS_antihelium3_x, GRAMS_antihelium3_y, GRAMS_antihelium3_lower, color="#FF7F7F", alpha=0.5, label="95% CL", zorder=2)

# Dark Matter (DM)
DM1, = plt.plot(DM_coogan_x, DM_coogan_y, label=f"Coogan et al.", linewidth=3, color="fuchsia", zorder=1)
DM2 = plt.fill_between(DM_Korsmeier_upper_x, DM_Korsmeier_lower_y_interp, DM_Korsmeier_upper_y, color="mediumblue", alpha=0.5, label="Korsmeier et al.", zorder=1)
DM3 = plt.fill_between(DM_Nan_Li_upper_x, DM_Nan_Li_lower_y_interp, DM_Nan_Li_upper_y, color="mediumpurple", alpha=0.5, label="Nan Li et al.", zorder=1)
DM4, = plt.plot(DM_MKachelrieβ_x, DM_MKachelrieβ_y, label=f"M. Kachelrieβ et al.", linewidth=3, color="darkviolet", zorder=1)
DM5, = plt.plot(DM_M_Winkler_x, DM_M_Winkler_y, label=f"M. Winkler & T. Linden", linewidth=3, color="midnightblue", zorder=1)

# Background (BKG)
BKG1, = plt.plot(BKG_Blum_x, BKG_Blum_y, linestyle='dashed', label=f"Blum et al.", linewidth=3, color="black", zorder=1)
BKG2, = plt.plot(BKG_Poulin_x, BKG_Poulin_y, linestyle='dashed', label=f"Poulin et al.", linewidth=3, color="dimgray", zorder=1)
BKG3, = plt.plot(BKG_A_Shukla_x, BKG_A_Shukla_y, linestyle='dashed', label=f"A. Shukla et al.", linewidth=3, color="silver", zorder=1)
BKG4, = plt.plot(BKG_M_Kachelrie_background_x, BKG_M_Kachelrie_background_y, linestyle='dashed', label=f"M. Kachelrieβ et al.", linewidth=3, color="gainsboro", zorder=1)

# Labels, title, and legend
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r"Particle Energy per nucleon [Gev/n]", fontsize=20)  # Increase font size for x-axis label
plt.ylabel(r'AntiHelium3 flux $[m^{2}\cdot s \cdot sr \cdot Gev/n]^{-1}$', fontsize=20)  # Increase font size for y-axis label
plt.xlim(0.1, 10)  # Adjust the limits to match your data
plt.ylim(1e-12, 1e-5)

plt.title("GRAMS Antihelium3 Sensitivity, 1 Event with 3 LDB Flight", fontsize=22, pad=20)  # Increase font size for the title

# Adjust tick font size
plt.tick_params(axis='both', labelsize=20)  # Increase font size for axis ticks

# Adjust tick parameters
plt.gca().tick_params(axis='both', which='major', length=15, width=1)  # Major ticks
plt.gca().tick_params(axis='both', which='minor', length=8, width=1)  # Minor ticks

# Create separate legends for datasets and ranges using bbox_to_anchor to place them separately
first_legend = plt.legend(handles=[GRAMS1, GRAMS2], loc='upper left', title="GRAMS sensitivity", bbox_to_anchor=(0.5, 1), fontsize=10)
second_legend = plt.legend(handles=[DM1, DM2, DM3, DM4, DM5], loc='upper left', title="Dark Matter", bbox_to_anchor=(0.75, 1), fontsize=10)
third_legend = plt.legend(handles=[BKG1, BKG2, BKG3, BKG4], loc='upper left', title="Background", bbox_to_anchor=(0.75, 0.77), fontsize=10)

# Add the legends to the plot
plt.gca().add_artist(first_legend)
plt.gca().add_artist(second_legend)

# Enable minor ticks
plt.minorticks_on()

# Add finer gridlines
plt.grid(which="major", linestyle='-', linewidth=0.8, color='gray', alpha = 0.2)  # Major gridlines
plt.grid(which="minor", linestyle='--', linewidth=0.5, color='lightgray', alpha = 0.3)  # Minor gridlines

# Show the plot
plt.show()
```
